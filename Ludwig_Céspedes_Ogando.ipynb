```python
import nltk
from nltk import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
nltk.download(["all"])
```

```python
stopwords = nltk.corpus.stopwords.words("english")
names = nltk.corpus.names.words()
```

```python
print(names)
print(stopwords)
```

```python
def get_tokens(series, reduce):

    vocabulary = []
    for comment in series:
        
        for idx, word in enumerate(nltk.word_tokenize(comment)):
            
            if not word.isalpha(): continue
           
            if word in stopwords: continue
            
            if word not in names: word = word.lower()
            
            vocabulary.append(reduce(word))
           
      

    return vocabulary
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
Buenasugenrencia = df[df["Sugerencia"]=="Buena sugenrencia"]["user_review"]
Malasugenrencia = df[df["Sugerencia"]=="Mala sugenrencia"]["user_review"]


positive_tokens = get_tokens(Buenasugenrencia, lemmatizer.lemmatize)
negative_tokens = get_tokens(Malasugenrencia, lemmatizer.lemmatize)
```

```python
positive_tokens
```

```python
negative_tokens 
```

```python
tafd = nltk.FreqDist(positive_tokens)
sefd = nltk.FreqDist(negative_tokens)
```

```python
top_200_positive = {word for word, count in tafd.most_common(200)}
top_200_negative = {word for word, count in sefd.most_common(200)}
```

```python
from statistics import mean

def extract_features(text):

    vocabulary = []
    for idx, word in enumerate(nltk.word_tokenize(text)):
        if not word.isalpha(): continue
        if word in stopwords: continue
        word = word.lower()
        word = lemmatizer.lemmatize(word)
        if word in top_200_negative or top_200_positive:
            vocabulary.append(word)

    fd = nltk.FreqDist(vocabulary)

    return fd
```

```python

Buenasugenrencia = df[df["Sugerencia"]=="Buena sugenrencia"]["user_review"].sample(200)
Malasugenrencia = df[df["Sugerencia"]=="Mala sugenrencia"]["user_review"].sample(200)

features = [
    (extract_features(review), "Postiva")
    for review in Buenasugenrencia
]
features.extend([
    (extract_features(review), "Negativa")
    for review in Malasugenrencia
])
```

```python
from random import shuffle

train_count = len(features)//2
shuffle(features)
classifier = nltk.NaiveBayesClassifier.train(features[:train_count])
classifier.show_most_informative_features(10)
```

```python
from sklearn.naive_bayes import (
    BernoulliNB,
    ComplementNB,
    MultinomialNB,
)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
classifiers = {
    "BernoulliNB": BernoulliNB(),
    "ComplementNB": ComplementNB(),
    "MultinomialNB": MultinomialNB(),
    "KNeighborsClassifier": KNeighborsClassifier(),
    "DecisionTreeClassifier": DecisionTreeClassifier(),
    "RandomForestClassifier": RandomForestClassifier(),
    "LogisticRegression": LogisticRegression(),
    "MLPClassifier": MLPClassifier(max_iter=1000),
    "AdaBoostClassifier": AdaBoostClassifier(),
}
```

```python
train_count = len(features) // 4
shuffle(features)

trained_classifiers = {}

for name, sklearn_classifier in classifiers.items():
     classifier = nltk.classify.SklearnClassifier(sklearn_classifier)
     classifier.train(features[:train_count])
     accuracy = nltk.classify.accuracy(classifier, features[train_count:])
     trained_classifiers[name] = classifier
     print(F"{accuracy:.2%} - {name}")
```

```python
import pickle
f = open('sentiment_classifier.pickle', 'wb')
pickle.dump(trained_classifiers["MultinomialNB"], f)
f.close()
```

```python
from sklearn.model_selection import train_test_split
x= df["user_review"].values
y = df["Sugerencia"].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.35, random_state = 4) 
print(len(x_train), len(x_test), len(y_train), len(y_test))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer(ngram_range = (1, 1), min_df = 0.1, max_df = 0.7, binary = False )
x_trainV = vect.fit_transform(x_train)
x_testV = vect.transform(x_test)
```

```python
x_df = pd.DataFrame(x_trainV.toarray(), columns = vect.get_feature_names_out())
print(x_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression

for name, sklearn_classifier in trained_classifiers.items():
    classifier = LogisticRegression() # Utilizamos la clase LogisticRegression como clasificador
    classifier.fit(x_trainV, y_train)
    y_pred = classifier.predict(x_testV)
    conf_mat = confusion_matrix(y_test, y_pred)
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.heatmap(conf_mat, annot=True, cmap="Blues", fmt="g", cbar=True)
    ax.set_xlabel("Predicted labels")
    ax.set_title("Confusion Matrix:{} ".format(trained_classifiers[name]))
    print("\n")
    plt.show()
```
